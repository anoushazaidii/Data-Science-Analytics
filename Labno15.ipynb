```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tests as t

%matplotlib inline

# Read in the datasets
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')
```

```python
movies
```

```python
ratings
```

```python
movie = movies.loc[:,["movieId","title"]]
movie.head(10)
```

```python
rating = ratings.loc[:,["userId","movieId","rating"]]
rating.head(10)
```

```python
# then merge movie and rating data
data = pd.merge(movie,rating)
```

```python
# now lets look at our data
data.head(10)
```

```python
data = data.iloc[:10000,:]
```

```python
# lets make a pivot table in order to make rows are users and columns are movies. And values are rating
pivot_table = data.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
pivot_table.head(10)
```

```python
movie_watched = pivot_table["Ace Ventura: Pet Detective (1994)"]
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
similarity_with_other_movies.head(20)
```

  Here we can conclude that we can recommend Things to Do in Denver When You're Dead (1995) to the people who watched Ace Ventura: Pet Detective (1994)
